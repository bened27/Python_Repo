```python
import pandas as pd
import numpy as np
import os
```

### Importing data

```python
path = r'/Users/jennyfevik/Desktop/IC Basket Analysis'
```

```python
ords_prods_merge = pd.read_pickle(os.path.join(path, 'Data', 'Prepared', 'orders_products_added_columns.pkl'))
```

### Aggregating Data from the Exercise

```python
# Creating and aggregating max order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

### Step 2 

```python
# Aggregating mean via grouped by
ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

The mean of a subset will always differ from that of an entire set, The entire set has all the data from all the departments, which means more data to factor in.

### Step 4

```python
# Setting loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order']>40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <=40) & (ords_prods_merge['max_order'] >10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <=10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

### Step 5

```python
# Basic statistics for loyalty_glag groups
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### Step 6

```python
# Creating mean price column
ords_prods_merge['mean_price'] = ords_prods_merge.groupby('user_id').agg({'prices': ['mean']})
```

```python
# creating spending flag
ords_prods_merge.loc[ords_prods_merge['mean_price'] <10, 'user_spending'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'user_spending'] = 'High spender'
```

```python
ords_prods_merge['user_spending'].value_counts()
```

### Step 7

```python
# creating the median for days since prior order
ords_prods_merge['median_days_since_order'] = ords_prods_merge.groupby('user_id').agg({'days_since_prior_order' :['median']})
```

```python
# creating user frequency flags
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] > 20, 'user_frequency'] = 'Non-frequent'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_order'] > 10) & (ords_prods_merge['median_days_since_order'] <= 20), 'user_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] <= 10, 'user_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge['user_frequency'].value_counts()
```
